```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fuu(x):
    return np.sqrt(1-x**2) + np.cbrt(x**2)

def bar(x):
    return -np.sqrt(1-x**2) + np.cbrt(x**2)
```

```python
x = np.linspace(-1, 1, 1000)
```

```python
y1 = fuu(x)
y2 = bar(x)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```
